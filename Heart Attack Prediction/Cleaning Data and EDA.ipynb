Columns Description:

Age: age of the person

Sex: gender/sex of person

CP: it stands for chest pain. it has 4 categories:
Typical angina (0)
Atypical angina (1)
Non-anginal pain (2)
Asymptomatic (3)

Trbtps: resting blood pressure (in mm Hg)

Chol: it provides the cholestrol level that is measured in mg/dl fetched via BMI sensor

Fbs: it stands for fasting blood sugar. if fbs>120 mg/dl then value=1 or else value=0

RestEcg: stands for esting electrocardiographic results. if:
Value 0: normal
Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria

Thalachh: max heatrate achieved by the person

Exng: exercise induced angina (1 = yes; 0 = no)

Oldpeak: ST depression induced by exercise relative to rest

Slp: the slope of the peak exercise ST segment.
0 = unsloping
1 = normal
2 = reversable effect

Caa: number of major vessels (0-3)

Thall: thalassemia
0 = null
1 = fixed defect
2 = normal
3 = reversable defect 

### Data Cleaning  ##

Data cleaning is a crucial step in the data preparation process, and its importance lies in the following key aspects:

Accuracy of Analysis:
Clean data ensures that the analysis and insights derived from it are accurate and reliable. Removing errors, inconsistencies, or outliers helps prevent misleading conclusions that may arise from flawed data.

Decision-Making Reliability:
Reliable data is essential for making informed decisions. Decision-makers rely on accurate information, and data cleaning contributes to the overall quality and trustworthiness of the data used in decision-making processes.

Model Performance in Machine Learning:
In machine learning, the performance of models heavily depends on the quality of the training data. Clean data ensures that the models are trained on accurate and representative information, leading to better predictive performance.

Consistency Across Datasets:
When integrating or combining datasets, ensuring consistency in data format, units, and values is critical. Data cleaning helps in harmonizing different sources of data, making them compatible for analysis.

Time and Cost Efficiency:
Cleaning the data upfront reduces the likelihood of errors and inconsistencies emerging later in the analysis process. This saves time and resources that would otherwise be spent addressing issues during or after analysis.

Enhanced Data Quality:
High-quality data is a valuable asset for any organization. Data cleaning contributes to improved data quality, which, in turn, enhances the overall trustworthiness and usefulness of the data for various applications.

Effective Visualization:
Clean data is more suitable for visualization purposes. Visualizations, such as charts and graphs, provide insights into trends and patterns in the data. Dirty data can distort these visual representations and lead to misinterpretations.

Regulatory Compliance:
In certain industries, there are regulations and standards regarding data quality and accuracy. Ensuring that the data meets these standards is essential for regulatory compliance and avoiding legal issues.

User Confidence:
Users, whether they are analysts, decision-makers, or consumers of the data, have more confidence in results derived from clean and well-organized data. It builds trust in the information provided.

Prevention of Bias and Unfairness:
Data cleaning helps in identifying and mitigating biases in the data. Biased data can lead to unfair or discriminatory outcomes, especially in machine learning applications.

### EDA ###

Exploratory Data Analysis (EDA) is crucial for several reasons in the process of understanding and preparing data for analysis. Here are some key reasons why EDA is important:

Understand the Structure of the Data:
EDA allows data analysts and scientists to understand the structure and organization of the dataset. It includes examining the number of observations, features, data types, and any missing values.

Identify Patterns and Trends:
EDA helps in identifying patterns, trends, and relationships within the data. Visualizations and statistical summaries aid in uncovering insights that may not be apparent at a glance.

Spot Anomalies and Outliers:
Outliers or anomalies in the data can significantly impact analysis and modeling. EDA helps identify these data points, enabling analysts to decide whether to exclude them, transform them, or investigate further.

Handle Missing Data:
EDA helps in understanding the extent of missing data and devising strategies to handle it. It includes exploring patterns of missingness and deciding on appropriate imputation methods.

Check for Data Quality:
Examining the quality of data is crucial for ensuring accuracy. EDA helps in identifying data entry errors, inconsistencies, or inaccuracies that need to be addressed through data cleaning processes.

Assess Data Distribution:
Understanding the distribution of numerical and categorical variables is essential for selecting appropriate statistical methods and models. EDA includes creating histograms, density plots, and other visualizations to assess data distribution.

Prepare for Statistical Modeling:
EDA is a precursor to statistical modeling. It helps analysts choose the most suitable statistical techniques, identify potential challenges, and make informed decisions about model assumptions.

Feature Selection and Engineering:
EDA aids in selecting relevant features for analysis and modeling. It also provides insights for creating new features through engineering, potentially improving the performance of machine learning models.

Improve Data Visualization:
Effective data visualization is a key part of EDA. It enables clear communication of insights to stakeholders and enhances the interpretability of results.

Enhance Hypothesis Testing:
Before performing hypothesis tests, it's crucial to understand the characteristics of the data. EDA helps formulate hypotheses, choose appropriate tests, and interpret results in the context of the data's characteristics.

Guide Exploratory Modeling:
In exploratory modeling, where the emphasis is on understanding the relationships between variables rather than predictive accuracy, EDA plays a pivotal role in guiding the initial modeling approach.

Communication and Reporting:
EDA results in visualizations and summaries that can be used in reports or presentations. It facilitates communication of findings to non-technical stakeholders and supports data-driven decision-making.

```python
#We use numerical features when used continous values
#Meaning values that can take any numeric value within a range
numerical_features = df[['age', 'trtbps', 'chol', 'thalachh', 'oldpeak']]

#We use categorical feautures when dealing with discretes values
#represent different categories or groups, 
#and the values are often non-numeric or represent specific codes
categorical_features = df[['sex', 'cp', 'restecg', 'exng', 'slp', 'caa', 'thall']]
```

This separation is essential because different types of features may require different types of analysis. 

Numerical features are typically used for correlation analysis, regression modeling, and other statistical methods that rely on continuous data. 

Categorical features, on the other hand, may involve techniques such as chi-square tests, analysis of variance (ANOVA), or they may be encoded for machine learning algorithms.

Separating them allows you to apply appropriate analysis methods to each group and create meaningful visualizations, such as correlation matrices and heatmaps, which can help in understanding the relationships within each category.

#### Boxplots #### 

##### "A box plot is a chart that shows data from a five-number summary including one of the measures of central tendency. It does not show the distribution in particular as much as a stem and leaf plot or histogram does. But it is primarily used to indicate a distribution is skewed or not and if there are potential unusual observations (also called outliers) present in the data set."  #####


- The box represents the interquartile range (IQR) and contains the central 50% of the data.

- The line inside the box represents the median.

- Whiskers extend to the minimum and maximum values within 1.5 times the IQR from the first and third quartiles.

- Points beyond the whiskers are considered potential outliers.

```python
plt.figure(figsize=(10, 8))
sns.heatmap(numerical_features.corr(), annot=True,  fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix for Numerical Features', fontsize=16)
plt.show()
```
