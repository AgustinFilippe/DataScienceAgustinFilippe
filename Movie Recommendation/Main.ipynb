```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('MovieDataset.xlsx')
```

```python
df = pd.DataFrame(data)
df['Duration (min)'] = df['Duration (min)'] / 10
df.head(10)
```

```python
import re
```

```python
def camel_case_split(s):
    return re.findall(r'[A-Z](?:[a-z]+|[A-Z]*(?=[A-Z]|$))', s)

df['Genre'] = df['Genre'].apply(lambda x: '/'.join(camel_case_split(x)))
```

```python
df.head(10)
```

```python
len(df)
```

```python
df.insert(0, 'id', df.index)
```

```python
df
```

```python
df.head(10)
```

```python
def sacar_numeros(arg):
    s = arg.lower()
    s = re.sub(r"\d+", "", s)
    return s
```

```python
sacar_numeros(" Me vi 30404 peliculas en 2 años")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(preprocessor = sacar_numeros,
                    min_df = 5,
                    max_df = 6278 )

bag_of_words = (cv.fit_transform(df['Summary']).toarray())

bag_of_words
```

```python
summary_columns = [tup[0] for tup in sorted(cv.vocabulary_.items(), key = lambda x: x[1])]

bag_of_words_df = pd.DataFrame(bag_of_words, columns = summary_columns, index = df['Title'])

bag_of_words_df.head()
```

```python
from itertools import combinations

def tokenizador_generos(string_generos):
    generos_separados = string_generos.split("/")
    resultado = []
    for tamaño in [1,2]:
        combs = ['Genre - ' + "/".join(sorted(tupla)) for tupla in combinations(generos_separados, r = tamaño)]
        resultado = resultado + combs
    return sorted(resultado)
```

```python
tokenizador_generos('Action/Crime/Drama')
```

```python
contador_generos = CountVectorizer(tokenizer = tokenizador_generos, token_pattern = None, lowercase = False)

contador_generos.fit(df['Genre'])

generos_bag_of_words = contador_generos.fit_transform(df['Genre']).toarray()

generos_bag_of_words
```

```python
columnas_genero = [tup[0] for tup in sorted(contador_generos.vocabulary_.items(), key = lambda x: x[1])]

bag_of_words_generos = pd.DataFrame(generos_bag_of_words, columns = columnas_genero, index = df['Title'])

bag_of_words_generos.head()
```

```python
bag_of_words_ambos = np.hstack((bag_of_words_df, bag_of_words_generos))

bag_of_words_ambos_df = pd.DataFrame(bag_of_words_ambos, columns = columnas_genero + summary_columns, index = df['Title'])

bag_of_words_ambos_df
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tf_idf = TfidfTransformer()
tf_idf_movies = tf_idf.fit_transform(bag_of_words_ambos_df).toarray()
tf_idf_movies
```

```python
tf_idf_movies = pd.DataFrame(tf_idf_movies, index=df['Title'], columns = columnas_genero + summary_columns)
```

```python
tf_idf_movies
```

```python
from sklearn.metrics.pairwise import cosine_similarity

c_similarity = cosine_similarity(tf_idf_movies)

recomendaciones = pd.DataFrame(c_similarity, index = df['Title'], columns = df['Title'])

recomendaciones
```

```python
np.fill_diagonal(recomendaciones.values, np.nan)  #Asi la peli no nos recomienda la propia pelicula

recomendaciones
```

```python
sort_movies = np.argsort((-c_similarity), axis = 1)

sort_movies
```

```python
c_similarity_sort = np.sort(-c_similarity, axis = 1)

c_similarity_sort
```

```python

def movie_recommendation(movie, k):
    # Obtén la fila correspondiente a la película
    movie_row = df[df['Title'] == movie].index[0]

    # Obtén las películas más similares y las similitudes correspondientes
    movie_sorted_sim = sort_movies[movie_row]
    lista_ordenada_sims = c_similarity_sort[movie_row]

    # Obtiene las primeras k películas
    top_k = movie_sorted_sim[:k]
    cosine_sim_top_k = lista_ordenada_sims[:k]

    # Crea un DataFrame con las películas más similares y sus similitudes
    top_k_df = df.loc[top_k].copy()
    top_k_df['Similaridad'] = cosine_sim_top_k

    return top_k_df
```

```python
df.drop(df.columns[0], axis=1, inplace=True)
```

```python
movie_recommendation('Iron Man 2',10)
```
