```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
import re
```

```python
url = 'https://www.imdb.com/search/title/?title_type=feature&sort=num_votes,desc'
options = Options()
options.headless = False
driver = webdriver.Chrome(options=options)
driver.get(url)
driver.maximize_window()
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.5'
}
res = requests.get(url, headers=headers)
html = res.content
soup = BeautifulSoup(html, 'html.parser')
```

```python
a = 0
dataset_movies = []
while True:
    movies = driver.find_elements(By.CLASS_NAME, 'ipc-metadata-list-summary-item')
    for i in range(a, len(movies)):
        
        movie = movies[i]

        time.sleep(1)
        #localizar y abrir la pestaña de informacion de pelicula
        info = movie.find_element(By.CLASS_NAME, 'sc-73c670dc-5')
        info.click()
        time.sleep(1)

        data_element = driver.find_element(By.CLASS_NAME, 'ipc-promptable-base__focus-lock')

        data_text = data_element.text

        # Split en los datos 
        lines = data_text.split('\n')

        # Extraer datos separados por las lineas
        try:
            title = lines[0]
        except Exception:
            title = ''
        
        try:
            year = lines[1][:4]
        except Exception:
            year = ''

        try:
            duration = int(lines[1][4:5])* 60 + int(lines[1][7:9])
        except Exception:
            try:
                duration = int(lines[1][4:5])* 60 + int(lines[1][7:8]) 
            except Exception:
                try:
                    duration = int(lines[1][4:5])* 60
                except Exception:
                    duration = '' 

        try:
            genre = lines[2]
        except Exception:
            genre = ''
            
        try:
            rating = lines[3]
        except Exception:
            rating = ''
            
        try:
            description = '\n'.join(lines[6:7])
        except Exception:
            description = ''
            
        try:
            director = lines[7]
        except Exception:
            director = ''
            
        

        # Para saber cuantas pelis scrapeadas vamos
        print(a , title)
        dataset_movies.append([title, year, duration, genre, rating, description, director])
        a += 1

        #localizar el boton de cerrar pestaña para iterar con las otras pels
        cerrar_info = driver.find_element(By.CLASS_NAME, 'ipc-promptable-base__close')
        cerrar_info.click()

        driver.execute_script("arguments[0].scrollIntoView();", movie)

        try:
            if a % 50 == 0:
                see_more_span = driver.find_element(By.CLASS_NAME, 'ipc-see-more__text')
                see_more_span.click()
                time.sleep(3)
        except Exception:
            continue
    
```
